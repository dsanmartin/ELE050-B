```python
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
```

# Datos

```python
X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Modelo

```python
model1 = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
```

```python
model1.score(X_test, y_test), model1.score(X_train, y_train)
```

# K-Fold cross validation

```python
model_cv = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(model_cv, X, y, cv=4)
```

```python
scores.mean()
```

# Ajuste de hiperpar√°metros

```python
hyperparams_grid = [
    #'C': [1, 10, 100, 1000, 10000],
    #'kernel': ['linear', 'rbf', 'poly', 'sigmoid']
    {'C': [1, 10, 100, 1000, 10000], 'kernel': ['linear']},
    {'C': [1, 10, 100, 1000, 10000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']}
]
```

```python
svc = svm.SVC()
grid_search = GridSearchCV(svc, hyperparams_grid, cv=4)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_, grid_search.best_params_
```

```python
grid_search.best_estimator_.score(X_test, y_test)
```

## Aleatorizado

```python
random_grid_search = RandomizedSearchCV(svc, hyperparams_grid, cv=4)
random_grid_search.fit(X_train, y_train)
```

```python
random_grid_search.cv_results_['params']
```
